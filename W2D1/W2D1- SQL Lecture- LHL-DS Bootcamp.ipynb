## Lets Analyze Tables using Postgresql and Instacart Data from Kaggle
[source: 'https://www.kaggle.com/c/instacart-market-basket-analysis/data' ]

The dataset for this competition is a relational set of files describing customers' orders over time. The goal of the competition is to predict which products will be in a user's next order. The dataset is anonymized and contains a sample of over 3 million grocery orders from more than 200,000 Instacart users. For each user, we provide between 4 and 100 of their orders, with the sequence of products purchased in each order. We also provide the week and hour of day the order was placed, and a relative measure of time between orders. For more information, see the blog post accompanying its public release.

   1. List total number of departments, then same for products (COUNT)
   <code>
    select count(distinct department_id) from departments;
    select count(distinct department_id) from products_small; #verify in other table
</code>

   2. List product_names in aisle 38 (single-table; WHERE)
   <code>
     select distinct(product_name) from products_small where aisle_id = 38;
   </code>
    
   3. List products  belong to department 1 and aisles 116 or 119 (single-table; WHERE with AND)
   <code>
    select distinct(product_name) FROM products_small where department_id =1 and aisle_id in (116,119);
   </code>
   
   4. List product names that belong to department 20 AND AISLE 96 and order them in ascending order of product ids
    <code>
    select distinct(product_name),product_id from products_small  where department_id=20 and aisle_id=96 order by product_id asc;
    </code>

  5. List all all departments with their product information (INNER JOIN)
  <code>
     select A.department,B.* from departments as A INNER JOIN products_small as B on A.department_id = B.department_id;
  </code>
  6. List all aisle_id's with more than 10 product_ids
  <code>
     select aisle_id, count(distinct product_id) as product_count from products_small group by aisle_id having count(distinct product_id) > 10;
  </code>
  7. List all departments along with how many aisles exist for each department (GROUP BY with INNER JOIN and COUNT)
  <code>
    select A.department,count(distinct B.aisle_id) as total_aisle_ids from departments as A INNER JOIN products_small as B on A.department_id = B.department_id group by A.department order by total_aisle_ids desc;
  </code>

## Why Databases
* Databases make it easy to store and retreive data
* They have structure and are rigid which does not allow data to be inserted only if it is in correct format.
* they follow acid properties:
    1. Atomicty: either all data gets inserted in a table or none
    2. Consistency: data is consistent before and after operations( inserting doesnt effect other records)
    3. Isolation: multiple transactions can occur simultaneously without causing an error.concurrency management is performed by the database
    4. Durbility: changes to database are permanent.

## Why SQL (history, rationale, etc)
* SQL is the language to communicate with databases.
* Databases need a protocol to communicate and that is SQL.
* SQL has data enforcement properties like Normalziation( not covered today.)
* schema design and data architecture can be performed using SQL.
* The SQL programming language was first developed in the 1970s by IBM researchers Raymond Boyce and Donald Chamberlin.

## Most often used DB schemes (star, snowflake...)
### Star Schema
1. Star schema is a top-down model.
2. Star Schema uses more space
3. It takes less time for the execution of queries.
4. It’s design is very simple.
5. It has less number of foreign keys.
6. It has high data redundancy.

### Snowflake Schema
1. it is a bottom-up model
2. It uses less space
3. it takes more time than star schema for the execution of queries.
4. it’s design is complex. So are the queries
5. it’s understanding is difficult.
6. Less no of foreign keys
7. Low data redundancy
