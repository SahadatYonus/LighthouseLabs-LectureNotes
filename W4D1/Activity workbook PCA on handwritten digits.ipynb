```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt
```

#### 1. download digits dataset using load_digits dataset available in sklearn. Load X and y from .data and .target 

```python
digits = load_digits()
X = digits.data
y = digits.target
```

#### 2. Perform PCA for 2 components on it

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
```

#### 3. Visualize the <code> reduced_X </code> output obtained after  PCA transformation by passsing it in the following function

```python
def visualize_2d_pca(original_X=None,X_reduced=None):
    print('Projecting %d-dimensional data to 2D' % original_X.shape[1])

    plt.figure(figsize=(12,10))
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, 
                edgecolor='none', alpha=0.7, s=40,
                cmap=plt.cm.get_cmap('nipy_spectral', 10))
    plt.colorbar()
    plt.show()
```

```python
visualize_2d_pca(X,X_reduced)
```

```python
def plot_original_and_components(row_no=20):
    plt.imshow(X[row_no].reshape((8, 8)), cmap='binary')
    plt.xticks([])
    plt.yticks([])
    plt.title('Source image')
    plt.show()

    fig, axes = plt.subplots(8, 8, figsize=(10, 10))
    fig.subplots_adjust(hspace=0.1, wspace=0.1)

    for i, ax in enumerate(axes.flat):
        pca = PCA(i + 1).fit(X)
        im = pca.inverse_transform(pca.transform(X[row_no].reshape(1, -1)))

        ax.imshow(im.reshape((8, 8)), cmap='binary')
        ax.text(0.95, 0.05, 'n = {0}'.format(i + 1), ha='right',
                transform=ax.transAxes, color='red')
        ax.set_xticks([])
        ax.set_yticks([])
```

#### 5. Pass the same number as above label from previous cell in the above function <code>plot_original_and_components</code> to look at how the PCA components look like and how much information they retain

```python
plot_original_and_components(40)
```
