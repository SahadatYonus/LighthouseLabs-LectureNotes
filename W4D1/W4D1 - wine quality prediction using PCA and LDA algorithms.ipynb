### Defining X and y for training and PCA analysis. In this wine dataset from 'http://archive.ics.uci.edu/ml/datasets/wine' the Customer Segment Column is not well defined but it is okay to use for our PCA operation

```python
X = df.iloc[:,0:13].values
y = df.iloc[:,13].values
```

```python
y
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=0)
```

#### Scaling the features:

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

#### Classification Algorithm Logistic Regression without PCA

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Confusion Matrix without PCA using LogReg

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print("This confusion matrix without PCA or LDA. ")
print(cm)
```

#### With PCA Logistic Regression Application

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) # we are reducing, but to how many components? 2.
X_trainPCA = pca.fit_transform(X_train)
X_testPCA = pca.transform(X_test)
```

```python
model = LogisticRegression(random_state = 0)
model.fit(X_trainPCA, y_train)
y_predPCA = model.predict(X_testPCA)
```

```python
cmPCA = confusion_matrix(y_test,y_predPCA)
print("This confusion matrix with PCA.")
print(cmPCA)
```

#### With LDA Logistic Regression Application

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 2) # Again choosed 2 component.
X_train_LDA = lda.fit_transform(X_train, y_train)
X_test_LDA = lda.transform(X_test)
```

```python
modelLDA = LogisticRegression(random_state = 0)
modelLDA.fit(X_train_LDA, y_train)
y_predLDA = modelLDA.predict(X_test_LDA)
cmLDA = confusion_matrix(y_test, y_predLDA)
print("This confusion matrix with LDA.")
print(cmLDA)
```

#### LDA performs better on this 14 column dataset. But PCA is not far behind. This makes sense because LDA seperates the classes or target varaibles and minimizes the scatter

#### Visualize LDA and y_test

```python
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.scatter(
    X_test_LDA[:,0],
    X_test_LDA[:,1],
    c=y_test,
    cmap='rainbow',
    alpha=0.7,
    edgecolors='b'
)
```
