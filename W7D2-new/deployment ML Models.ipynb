### REST API
[source: 'https://www.smashingmagazine.com/2018/01/understanding-using-rest-api/' ]

What is REST API?

* An API is an application programming interface. It is a set of rules that allow programs to talk to each other. The developer creates the API on the server and allows the client to talk to it.
* REST determines how the API looks like. It stands for “Representational State Transfer”. 
* It is a set of rules that developers follow when they create their API. One of these rules states that you should be able to get a piece of data (called a resource) when you link to a specific URL.


### The Anatomy Of A Request
* It’s important to know that a request is made up of four things:

1.The endpoint

2.The method

3.The headers

4.The data (or body)

#### The endpoint (or route) is the url you request for

The endpoint (or route) is the url you request for

<code>root-endpoint/?</code>

<code>https://api.github.com</code>

### The Method

#### The method is the type of request you send to the server. You can choose from these five types below:

1.GET - Used to get resource from server

2.POST - Used to create new resource on server

3.PUT/PATCH - update resource on server

4.DELETE - delete a resource on the server

#### These methods provide meaning for the request you’re making. They are used to perform four possible actions: Create, Read, Update and Delete (CRUD).
